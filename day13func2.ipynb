```python
def details(name, age, city):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"City: {city}")
    
print("--- Positional Call ---")
details("Alice", 22, "New York")


print("--- Keyword Call ---")
details(city="London", name="Bob", age=25)
```

```python
def calculate(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "divide":
        if b == 0:
            return "Cannot divide by zero!"
        return a / b
    else:
        return "Invalid operation"

print(calculate(10,5,"add"))
```

```python
def order(item, quantity, price):
    total_cost = quantity * price
    print(f"Item: {item}")
    print(f"Quantity: {quantity}")
    print(f"Price per unit: ${price}")
    print(f"Total Bill: ${total_cost}")
order("Laptop", price=1200, quantity=2)
```

```python
def add_all(*args):
    return sum(args)
add_all(1,2,3,4,5,6,7,8)
```

```python
def print_all(*args):
    for item in args:
        print(item)
print_all("apple",10,7.8)
```

```python
def find_max(*args):
    return max(args)
print(find_max(1,2,3,45,67))
```

```python
def concatenate(*args):
    return " ".join(args)
print(concatenate("hello","world","python"))
```

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result
print(multiply(1,2,3,4))
```

```python
def calculate_average(*args):
    return sum(args)/len(args)
print(calculate_average(60,50,89,90))
```

```python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
student_info(name="akshitha",age=21,course="ece")
```

```python
def create_profile(**kwargs):
    for key, value in kwargs.items():
        print(key,":",value)
create_profile(
    name="akshitha",
    age=21,
    city="hyd",
    course="ece",
    college="sr university")
```

```python
def calculate_total(**kwargs):
    return sum(kwargs.values())
print(calculate_total(rice=120,milk=60,bread=40,eggs=80))
```

```python
def save_data(**kwargs):
    print("saving data:")
    for key,value in kwargs.items():
        print(key,":",value)
save_data(name="akshitha",age=21,city="hyd")
    
```
