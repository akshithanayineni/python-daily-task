```python
class Student:
    total_students = 0
    def __init__(self, marks):
        self.__marks = marks
        self.status = "Passed"
        Student.total_students += 1
    def get_marks(self):
        return self.__marks
    def set_marks(self, marks):
        if min(marks) < 0 or max(marks) > 100:
            print("Invalid")
        elif sum(marks)/5 < sum(self.__marks)/5:
            print("Average cannot decrease")
        else:
            self.__marks = marks
            if min(marks) < 40:
                self.status = "Failed"
s = Student([80, 70, 90, 85, 75])
print(s.get_marks())
s.set_marks([85, 80, 95, 90, 88])
print(s.get_marks())
print(s.status)
print(Student.total_students)
```

```python
class Product:
    total_revenue = 0
    def __init__(self, price):
        self.__price = price
        self.__discount = 0
        self.original_price = price
    def get_price(self):
        return self.__price
    def set_price(self, price):
        if price < self.original_price * 0.8:
            print("Price cannot be less than 80% of original price")
        else:
            self.__price = price
            if price > 5000:
                self.__discount = 15
            Product.total_revenue += price
    def get_discount(self):
        return self.__discount
p = Product(10000)
print(p.get_price())
p.set_price(6500)
print(p.get_price())
print(p.get_discount())
print( Product.total_revenue)
```

```python
class User:
    senior_citizen = False
    def __init__(self, password, age):
        self.__password = password
        self.__age = age
    def get_age(self):
        return self.__age
    def set_password(self, password):
        if len(password) >= 8 and any(ch.isdigit() for ch in password):
            self.__password = password
        else:
            print("Invalid Password")
    def set_age(self, age):
        if age < self.__age:
            print("Age cannot be decreased")
        else:
            self.__age = age
            if age >= 60:
                User.senior_citizen = True
    def update(self, password, age):
        self.set_password(password)
        self.set_age(age)
u = User("12345", 60)
u.update("123456", 70)
print(u.get_age())
print(User.senior_citizen)
```
