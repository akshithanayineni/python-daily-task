Take three sides of a triangle. First check if they form a valid triangle, then check
if it is Equilateral, Isosceles, or Scalene.

```python
a=int(input())
b=int(input())
c=int(input())
if a+b>c or b+c>a or c+a>b:
    print("valid triangle")
    if a==b==c:
        print("equilateral triangle")
    else:
       if a==b!=c or a!=b==c:
           print("isoceles")
       else:
            print("scalene")
else:
    print("invalid triangle")
```
