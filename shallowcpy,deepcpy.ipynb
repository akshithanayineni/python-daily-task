```python
import copy
def detect_copy(original,copied):
    copied[2][0]=100
    if original==copied:
        print("shallow")
    else:
        print("deep")
original=[1,2,[3,4]]
shallow=copy.copy(original)
detect_copy(original,shallow)
original=[1,2,[3,4]]
deep=copy.deepcopy(original)
detect_copy(original,deep)

    
```

```python
class Bankaccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(self.balance)
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount!")
        elif amount <= self.balance:
            self.balance -= amount
            print(f"Withdraw {amount}. Remaining Balance: ₹{self.balance}")
        else:
            print(f"Transaction failed! Insufficient balance. Available Balance: ₹{self.balance}")

account = Bankaccount(account_holder="aksh", balance=2000)
account.deposit(500)
account.withdraw(300)
account.withdraw(2000)
```

```python
class StudentAttendance:
    def __init__(self, total_classes, attended_classes):
        self.total_classes = total_classes
        self.attended_classes = attended_classes
    def mark_attendance(self, number):
        if self.attended_classes + number <= self.total_classes:
            self.attended_classes=self.attended_classes+number
            percentage=(self.attended_classes / self.total_classes)*100
            print("attendance updated successfully")
            print("attendance percentage:",percentage,"%")
        else:
            print("attendance cannot exceed ")
student = StudentAttendance(100,89)
student.mark_attendance(10)
            
```

```python
class BusReservation:
    def __init__(self, bus_name, total_seats, booked_seats):
        self.bus_name = bus_name
        self.total_seats = total_seats
        self.booked_seats = booked_seats
    def reserve(self, number):
        if self.booked_seats + number <= self.total_seats:
            self.booked_seats += number
            remaining_seats = self.total_seats - self.booked_seats
            print("reservation successful")
            print("remaining seats:",remaining_seats)
        else:
            print("not possible")
bus=BusReservation("express",50,40)
bus.reserve(5)
```

```python
class OnlineShoppingCart:
    def __init__(self, customer_name, wallet_balance):
        self.customer_name = customer_name
        self.cart_total = 0
        self.wallet_balance = wallet_balance

    def add_item(self, price):
        self.cart_total = self.cart_total + price
        print("Item added")
        print("Cart Total:", self.cart_total)

    def remove_item(self, price):
        if self.cart_total - price >= 0:
            self.cart_total = self.cart_total - price
            print("Item removed")
            print("Cart Total:", self.cart_total)
        else:
            print("Cart total cannot become negative")

    def checkout(self):
        if self.wallet_balance >= self.cart_total:
            self.wallet_balance = self.wallet_balance - self.cart_total
            self.cart_total = 0
            print("Payment successful")
            print("Wallet Balance:", self.wallet_balance)
            print("Cart Total:", self.cart_total)
        else:
            print("Insufficient wallet balance")


cart = OnlineShoppingCart("Akshitha", 5000)

cart.add_item(1000)
cart.add_item(1500)
cart.remove_item(500)
cart.checkout()
```
